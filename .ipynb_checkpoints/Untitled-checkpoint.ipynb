## Project Overview


In general, I would like to leverage the California Health Interview Survey (CHIS) dataset to identify interesting insights between individual patient characteristics and health outcomes.  The CHIS, conducted by UCLA, is an annual survey of ~20k individuals where 300+ data points (i.e. fields) are collected on each individual.  The general categories of data include: demographics; general health and behaviors with specific questions around heart disease, diabetes, and asthma; mental health; health utilization; health insurance; and employment and socioeconomic.  While there are a number of questions that could be pursued with this dataset (e.g. what are the indicatators of heart disease, diabetes, mental health conditions, etc.), I chose to focus on identifying individuals who are 'high utilizers' of healthcare.  Specifically, I am aiming to predict the individuals who will visit a doctor's office greater than 12 times in a year.  These individuals are likely high cost individuals (given the high health care utilization); so it could make sense for a health system or payer to focus additional care management resources around managing these specific individuals.  If one could identify the individuals who are going to be high utilizers of healthcare and put in place care management intiatives focused on these individuals; a health system or payer might be able to simulaneously decrease overall costs and increase patient quality of care.  Given this use case, I plan to use select fields from the 2009-2014 CHIS survey data to predict which individuals have greater than 12 doctors visits in a specific year.  Note that I am pursuing this as a classification problem, since the decision to put a care manager on a patient most closely resembles a yes or no.  Based on this real world decision, in this use case, I care less about predicting the exact number of doctors visits and more about identifying high utilizers.

## Project Caveats

I want to list a number of caveats before getting started so that any readers are aware of the limitation of the analyses described throughout this project 

* Likely most importantly, the survey is not longitudinal and most questions relate to events that occured in the last twelve months.  This is very important when extrapolating out learnings from the CHIS dataset to real life environments.  As a specific example, if an individual visited the ER and developed heart disease in a given year; they might have a lot of doctor's visits in the CHIS dataset.  However, in the CHIS dataset, we don't know the order in which these events occured.  If we found that ER visits and heart disease are key indicators of a high number of doctor's visits, this would not help us in the real world predict hospital visits for someone who (i) hasn't visited the ER in the past but will in the coming year for some reasons or (ii) does not yet have heart disease but will develop it in the next year.  
* CHIS provides the weights of individual records to account for the fact that the individuals they surveryed aren't perfectly representative of the overall Californian population.  I do not take into account the sample weights provided by CHIS to adjust for survey sample bias.  Since I am just trying to get directional feature importance; I do not think this is critical at this stage, but it could be a future step.  Importantly, if one wanted to use this data set as a benchmark to compare 'population level data' to other data sets (e.g. what is average healthcare utilization, payer mix, etc.), one would want to take the sample weights into account to ensure the survey sample is more representative of the CA population as a whole and not just the surveyed population



## Background on Dataset

**General Backgroud on CHIS Dataset**

The California Health Interview Survey (CHIS) is an ongoing cross-sectional survey of Californiaâ€™s population.  The specific questions that are asked to individuals are set for a two year period and subsequently tweaked every two years.  UCLA has conducted the survey for over a decade, with results now released every year (note that historically, results were only released every two years).  Approximately 20k individuals are surveyed each year.  While each row in a survey represents a specific individual's response, the survey is not longitudinal because there is no way to tie if an individual answered a survey in multiple years (and most survey responses in a given year are likely taking the survey for the first time).  I provide the questionnaires and data dictionaries for the 2009-2014 surveys in the "Questionnaires and Data Dictionaries" folder.  Additional details on the CHIS methodological can be found online (http://healthpolicy.ucla.edu/chis/design/Pages/methodology.aspx).  

**Specific Backgroud on How I am Using the CHIS Dataset**

I first looked at the most recent survey results from 2013 and 2014.  While there are over 400 fields in these datasets (the fields are identical since the same questions were asked in 2013 and 2014), I went through and selected ~50 fields that I believed were most relevant to the exercise of predicting the number of doctors visits in a given year.  I did not start with all fields mostly because of the repetitiveness of certain fields and the amount of data cleaning required for each field.  In addition to the 2013 and 2014 datasets, I wanted to use as much of the historical CHIS data as possible, as I believed having more data would help strengthen the  generalizability of specific models (all else being equal).  While the CHIS datasets go back to the early 2000's, the issue I ran into was the fact that the survey questions are tweaked every two years.  Given this, I found ~35 fields that were fairly consistent from 2009-present; so I used these 35 fields and concatenated 2009 through 2014 data, giving me ~130k indvidual responses.  I did not pull data prior to 2009 as the survey changed quite a bit between the 2007 and 2009 survey, so I would have had to majorly reduce the feature set.

```python
'''look at ROC_AUC curve'''
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(init=None, learning_rate=0.1, loss='deviance',
              max_depth=2, max_features=None, max_leaf_nodes=None,
              min_samples_leaf=33, min_samples_split=85,
              min_weight_fraction_leaf=0.0, n_estimators=250,
              random_state=1, subsample=1.0, verbose=0, warm_start=False)
clf.fit(X_train_scaled, y_train)
probs = clf.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, thresholds = metrics.roc_curve(y_test, probs)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
```
